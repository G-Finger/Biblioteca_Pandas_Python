O método drop() da biblioteca Pandas permite remover uma ou mais linhas e/ou colunas de um DataFrame. Essa remoção pode ser realizada de forma permanente no DataFrame original. É possível especificar os índices das linhas e/ou colunas a serem removidos, bem como o eixo em que a remoção deve ocorrer.

Suponha que temos o seguinte DataFrame, salvo em uma variável chamada dados:

Código para criação do DataFrame:

```python
import pandas as pd

dados = pd.DataFrame([['Feira', 'Cebola', 2.5],
                        ['Mercado', 'Cebola', 1.99],
                        ['Supermercado', 'Cebola', 1.69],
                        ['Feira', 'Tomate', 4],
                        ['Mercado', 'Tomate', 3.29],
                        ['Supermercado', 'Tomate', 2.99],
                        ['Feira', 'Batata', 4.2],
                        ['Mercado', 'Batata', 3.99],
                        ['Supermercado', 'Batata', 3.69]],
                        columns = ['Local', 'Produto', 'Preço'])
dados
```

Qual dos códigos abaixo poderia ser utilizado para remover de forma permanente as linhas onde, na coluna "Local", temos o "Supermercado"?

```python
 dados.drop([2, 5, 8], axis=0, inplace=True)
 dados
```

Para praticar os métodos aprendidos no decorrer dessa aula e também aprender novos, vamos realizar alguns tratamentos e seleções utilizando um arquivo csv diferente:

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
pd.read_csv(url)
dados_alunos = pd.read_csv(url)
dados_alunos
```

1) Verifique se a base de dados possui dados nulos e, caso tenha, realize o tratamento desses dados nulos da forma que achar mais coerente com a situação.

```python
dados_alunos.isnull().sum()
```

```python
dados_alunos.Notas.mean()
```

Perceba que há 6 notas que não são válidas. Vamos colocar zero nelas. .

```python
dados_alunos['Notas'] = dados_alunos['Notas'] .fillna(0)
dados_alunos
```

```python
dados_alunos.Notas.mean() # percebam que ela diminuiu o valor da média, porque ele considera agora as notas dos alunos como zero.
```

2) Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados.

```python
registros_a_remover = dados_alunos.query("Nome == 'Alice' | Nome == 'Carlos'").index
registros_a_remover
dados_alunos.drop(registros_a_remover, axis = 0, inplace=True) # com zero removemos linhas, com 1 removemos colunas
dados_alunos
```

3) Aplique um filtro que selecione apenas os alunos que foram aprovados.

```python
alunos_aprovados = dados_alunos.query("Aprovado == True")
alunos_aprovados
```

4) Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv".

```python
alunos_aprovados.to_csv('alunos_aprovados_csv', index=False, sep=';')
```

```python
pd.read_csv('alunos_aprovados_csv') # arquivo com os alunos aprovados.
```

Importar o seguinte arquivo para resolução dos exerícios:



```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
pd.read_csv(url)
dados_alunos = pd.read_csv(url)
dados_alunos
```

1) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.

```python
dados_alunos['Extrinhas'] = dados_alunos['Notas'] * 0.4
dados_alunos
```

2) Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.

```python
dados_alunos['Notas_Finais'] = dados_alunos['Notas'] + dados_alunos['Extrinhas']
dados_alunos
```

3) Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:

True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6)

False: caso o aluno esteja reprovado (nota final deve ser menor que 6).

```python
dados_alunos['Aprovado_Final'] = dados_alunos['Notas_Finais'].apply(lambda x:True if x>6 else False )
dados_alunos
```

4) Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.

```python
dados_alunos.query('Aprovado != Aprovado_Final & Aprovado_Final == True')
```

```python
alunos_aprovados_corrigido = dados_alunos.query('Aprovado != Aprovado_Final & Aprovado_Final == True')
alunos_aprovados_corrigido
```
