# Manipulando os dados

## Criando colunas numéricas

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep = ';')
dados.head() # estamos importanto a base novamente para trabalharmos com toda a base e não somente uma parte.
```

```python
dados['Valor_por_mes'] = dados['Condominio'] + dados['Valor']
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes']*12 + dados['IPTU']
dados.head()
```

## Criando colunas categóricas

```python
dados['Descrição'] = dados['Tipo'] + ' em ' + dados['Bairro'] # colocamos o 'em' para melhorar a apresentação.
dados.head()
```

```python
dados['Descrição'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'].astype(str) + ' quartos' + ' e ' + dados['Vagas'].astype(str)+ ' vagas de garagem.'
dados.head()
#colocamos o astype(str) para converter um valor inteiro para um valor do tipo string.
```

```python
dados['Possui_suites'] = dados['Suites'].apply(lambda x:"Sim" if x>0 else "Não") # criamos uma coluna em que mostra com SIM os imóveis que há suites
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
